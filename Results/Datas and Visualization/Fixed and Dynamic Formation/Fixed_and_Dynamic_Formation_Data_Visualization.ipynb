```python
import pandas as pd
import numpy as np
from bokeh.plotting import show , output_file, figure
from bokeh.models import Span, Legend, ColumnDataSource, NumeralTickFormatter
import csv
```

```python
x1 = pd.read_csv('x1.csv')                    # Importing Data.
x2 = pd.read_csv('x2.csv')
x3 = pd.read_csv('x3.csv')
x4 = pd.read_csv('x4.csv')
xc = pd.read_csv('xc.csv')

y1 = pd.read_csv('y1.csv')
y2 = pd.read_csv('y2.csv')
y3 = pd.read_csv('y3.csv')
y4 = pd.read_csv('y4.csv')
yc = pd.read_csv('yc.csv')
```

```python
output_file('Dynamic_Formation.html')                    # Creating Output.
xc3 = xc.to_numpy()                         # Preparing data for plotting.
y11 = y1.to_numpy()
x11 = x1.to_numpy()
y22 = y2.to_numpy()
x22 = x2.to_numpy()
y33 = y3.to_numpy()
x44 = x4.to_numpy()
x33 = x3.to_numpy()
y44 = y4.to_numpy()
yc3 = yc.to_numpy()
                                           # Drawing a figure on output.
p = figure(title='Dynamic Formation of four Agents',width=1500, height=800 ,toolbar_location=None, tools="",background_fill_color="#fafafa")
 
p.line(xc3[:,0], yc3[:,0], color='red', legend_label="Trajectory")      # Lining the Trajectory on the figure.

p.legend.location='top_left'              # Setting detailes of the figure.
p.xgrid.grid_line_color=None
p.axis.minor_tick_line_color=None
p.yaxis.axis_line_color=None
p.xaxis.axis_line_color='black'
p.axis.major_tick_line_color=None                                                 
p.outline_line_color = None  
show(p)
```

```python
a = [0 ,5 ,10, 15]                       # Plotting Dynamic formation Datas.

for i in a:
    yy1 = y11[15000+i*3000]              # We plot Agent #1 with hexes, Agent #2 with squares, 
    yy2 = y22[15000+i*3000]              #  Agent #3 with diamonds, Agent #4 with inverted triangles,
    yy3 = y33[15000+i*3000]              #  and the Virtual Leader with circles.
    yy4 = y44[15000+i*3000]
    yyc = yc3[15000+i*3000]
    xx1 = x11[15000+i*3000]
    xx2 = x22[15000+i*3000]
    xx3 = x33[15000+i*3000]
    xx4 = x44[15000+i*3000]
    xxc = xc3[15000+i*3000]
    

    p.hex(xx1[0],yy1[0], size=15)
    p.square(xx2[0],yy2[0], size=15, color='orange', fill_color='silver')
    p.diamond(xx3[0],yy3[0], size=16,color='darkred' , fill_color='red')
    p.inverted_triangle(xx4[0],yy4[0], size=16, color='blue', fill_color='silver' )
    p.line([xx1[0],xx2[0],xx3[0],xx4[0],xx1[0]], [yy1[0],yy2[0],yy3[0],yy4[0],yy1[0]],color='black', line_dash="4 4")
    p.circle(xxc[0], yyc[0], size=6, color="navy", alpha=0.5)
    

    
yyc = yc3[0]
xxc = xc3[0]    
p.circle(xxc[0], yyc[0], size=6, color="navy", alpha=0.5)
yyc = yc3[80000]
xxc = xc3[80000]    
p.circle(xxc[0], yyc[0], size=6, color="navy", alpha=0.5,legend_label="Virtual Leader")

p.line([x11[0][0],x22[0][0],x33[0][0],x44[0][0],x11[0][0]], [y11[0][0],y22[0][0],y33[0][0],y44[0][0],y11[0][0]],color='darkcyan', line_dash="4 4")
p.line([x11[80000][0],x22[80000][0],x33[80000][0],x44[80000][0],x11[80000][0]], [y11[80000][0],y22[80000][0],y33[80000][0],y44[80000][0],y11[80000][0]],color='fuchsia', line_dash="4 4")

p.hex(x11[0][0],y11[0][0], size=15, legend_label="Agent #1")
p.square(x22[0][0],y22[0][0], size=15, color='orange', fill_color='silver', legend_label="Agent #2")
p.diamond(x33[0][0],y33[0][0], size=16, color='darkred' , fill_color='red' , legend_label="Agent #3")
p.inverted_triangle(x44[0][0],y44[0][0], size=16, color='blue', fill_color='silver',fill_alpha=0.2, legend_label="Agent #4")


p.hex(x11[80000][0],y11[80000][0], size=15, legend_label="Agent #1")
p.square(x22[80000][0],y22[80000][0], size=15, color='orange', fill_color='silver', legend_label="Agent #2")
p.diamond(x33[80000][0],y33[80000][0], size=16, color='darkred' , fill_color='red' ,legend_label="Agent #3")
p.inverted_triangle(x44[80000][0],y44[80000][0], size=16, color='blue', fill_color='silver', legend_label="Agent #4")

show(p)

```
